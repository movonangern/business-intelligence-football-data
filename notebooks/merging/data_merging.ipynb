### Merge of all Dataframes

```python
# Pfad zum Verzeichnis, in dem sich Ihre CSV-Dateien befinden
path = '/Users/movonangern/Library/Mobile Documents/com~apple~CloudDocs/IT/Programming/Business Intelligence/business-intelligence-football-data/data/raw/data/FBdata'

# Liste aller CSV-Dateien, die zusammengeführt werden sollen, 
# 'soccer_dataframe_2324.csv' wird speziell behandelt
csv_files = [file for file in os.listdir(path) if file.endswith('.csv')]

# Leere Liste für DataFrames initialisieren
dataframes_list = []

# Durchlaufen der Liste der CSV-Dateien
for file in csv_files:
    # CSV-Datei einlesen
    df = pd.read_csv(os.path.join(path, file))
    dataframes_list.append(df.iloc[2:])

# Alle DataFrames in der Liste zusammenführen
final_df = pd.concat(dataframes_list, ignore_index=True)

# Pfad, wo die kombinierte CSV-Datei gespeichert werden soll
output_file = '/Users/movonangern/Library/Mobile Documents/com~apple~CloudDocs/IT/Programming/Business Intelligence/business-intelligence-football-data/data/merged/soccerdata.csv'
final_df.to_csv(output_file, index=False)

# Pfad zur neuen CSV-Datei zurückgeben
print(f'Der kombinierte DataFrame wurde gespeichert als: {output_file}')

```

### Setting new Column-Names

Our column names need to be merged from the first and second row of our original dataframe. We do that by setting new column names manually.


```python
new_column_names = ["league", "season", "game", "team", "player", "jersey_number", "nation", "pos", "age", "min", "Gls", "Ast", "PK", "PKatt", "Sh", "SoT", "CrdY", "CrdR", "Touches", "Tkl", "Int", "Blocks", "xG", "npxG", "xAG", "SCA", "GCA", "Cmp", "Att", "Cmp%", "PrgP", "Carries", "PrgC", "Att", "Succ", "game_id"]
len(new_column_names)
```

```python
df_soccerdata.columns = new_column_names
```

```python
df_soccerdata = df_soccerdata.iloc[2:]
```

```python
def extract_date(game_info):
    # Extrahiere das Datum basierend auf dem bekannten Format 'YYYY-MM-DD'
    return game_info.split()[0]  # Teilt den String am Leerzeichen und nimmt den ersten Teil (das Datum)

# Anwenden der Funktion auf jede Zeile in der Spalte 'game' und Speichern des Ergebnisses in einer neuen Spalte 'date'
df_soccerdata['date'] = df_soccerdata['game'].apply(extract_date)
```
