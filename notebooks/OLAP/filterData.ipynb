```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Verbindung zur Datenbank herstellen
engine = create_engine("mysql+mysqlconnector://root:root@localhost:3306/football_db", echo=False)
```

```python
query_appearances = "SELECT * FROM appearances WHERE date >= '2021-01-01';"
df_appearances = pd.read_sql_query(query_appearances, engine)
df_appearances.head(10)
```

```python
df_appearances.shape
```

```python
query_club_games = "SELECT * FROM club_games WHERE game_id IN (SELECT game_id FROM games WHERE season >= 2021);"
df_club_games = pd.read_sql_query(query_club_games, engine)
df_club_games.head(10)
```

```python
df_club_games.shape
```

```python
query_clubs = "SELECT * FROM clubs;"
df_clubs = pd.read_sql_query(query_clubs, engine)
df_clubs.head(10)
```

```python
df_clubs.shape
```

```python
query_competition = "SELECT * FROM competition;"
df_competition = pd.read_sql_query(query_competition, engine)
df_competition.head(10)
```

```python
df_competition.shape
```

```python
query_game_events = "SELECT * FROM game_events WHERE game_id IN (SELECT game_id FROM games WHERE season >= 2021);"
df_game_events = pd.read_sql_query(query_game_events, engine)
df_game_events.head(10)
```

```python
df_game_events.shape
```

```python
query_game_lineups = "SELECT * FROM game_lineups WHERE game_id IN (SELECT game_id FROM games WHERE season >= 2021);"
df_game_lineups = pd.read_sql_query(query_game_lineups, engine)
df_game_lineups.head(10)
```

```python
df_game_lineups.shape
```

```python
query_games = "SELECT * FROM games WHERE season >= 2021;"
df_games = pd.read_sql_query(query_games, engine)
df_games.head(10)
```

```python
df_games.shape
```

```python
query_player_valuations = "SELECT * FROM player_valuations WHERE date >= '2021-01-01';"
df_player_valuations = pd.read_sql_query(query_player_valuations, engine)
df_player_valuations.head(10)
```

```python
df_player_valuations.shape
```

```python
query_players = "SELECT * FROM players;"
df_players = pd.read_sql_query(query_players, engine)
df_players.head(10) 
```

```python
df_players.shape
```

```python
import os

# Erstelle den Ordner, wenn er nicht existiert
output_folder = "C:/Users/Richa/Desktop/business-intelligence-football-data/data/OLAP data"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Speichere jeden DataFrame als CSV in dem Ordner
df_appearances.to_csv(f"{output_folder}/df_appearances.csv", index=False)
df_club_games.to_csv(f"{output_folder}/df_club_games.csv", index=False)
df_clubs.to_csv(f"{output_folder}/df_clubs.csv", index=False)
df_competition.to_csv(f"{output_folder}/df_competition.csv", index=False)
df_game_events.to_csv(f"{output_folder}/df_game_events.csv", index=False)
df_game_lineups.to_csv(f"{output_folder}/df_game_lineups.csv", index=False)
df_games.to_csv(f"{output_folder}/df_games.csv", index=False)
df_player_valuations.to_csv(f"{output_folder}/df_player_valuations.csv", index=False)
df_players.to_csv(f"{output_folder}/df_players.csv", index=False)

```
